### What is OS?
os helps you work with your computer's file system.
- What folder am I currently in? os.getcwd()
- What's inside a folder? os.listdir("folder")
- Can I make a new folder? os.makedirs("newfolder", exist_ok=True)

```python
import os
import pandas as pd

folder = "../State Datasets" 
# we're currently in scrapers
# go up one folder
# go to state datasets
```

```python
os.listdir(folder)
# folder has all csv files + more!
```

```python
# getting all csv filenames in a list through list comprehension

csv_files = [
    file for file in os.listdir(folder) #os.listdir(folder) means show me everything inside this folder
    if file.endswith(".csv") #there will be all kinds of files like DS_Store etc. Keep only csv files.
]

csv_files
```

```python
# read every csv

from pandas.errors import EmptyDataError

all_dfs = []

for file in csv_files:
    filepath = os.path.join(folder, file)

    try: #I'm going to attempt something.
        df = pd.read_csv(filepath) #if it works, then it skips the except block entirely and goes to all_dfs.append(df)
    except EmptyDataError:
        print(f"Skipping empty file: {file}")
        continue

    all_dfs.append(df)
```

```python
national = pd.concat(all_dfs, ignore_index=True)
national.to_csv("../national.csv", index=False)
```
