### Creating Prompt

```python
def make_prompt(complaint, complaint_category):
    return f""" 

You are coding an animal-welfare complaint for a data-journalism project.

These are allegations made in complaints. Do not treat them as verified facts.

Use only:

1. the complaint text; and
2. the existing complaint_category supplied below.

Do not infer facts that are not clearly stated.

--------------------------------------------------
ELIGIBILITY OF COMPLAINTS TO CLASSIFY
--------------------------------------------------

Complete this classification only when complaint_category is one of:

- harm_or_threat_to_dogs
- harm_or_threat_to_dogs_and_feeders

For every other complaint_category, return "Not Applicable" for every output field.

--------------------------------------------------
MULTI-LABEL HARM CLASSIFICATION
--------------------------------------------------

Rules:

- For eligible complaints, assign either 1 or 0 to each harm category.
- A complaint may receive 1 for more than one category.
- Assign 1 only when the category is clearly alleged in the complaint.
- Assign 0 when it is not alleged.
- Threatened, attempted and completed acts are coded together under the relevant harm category.

Categories:

1. physical_cruelty

Assign 1 for alleged deliberate physical cruelty or an explicit threat of physical cruelty toward dogs.

Includes:

- deliberate killing
- poisoning
- beating
- kicking
- stabbing
- burning
- torture
- deliberate injury
- forceful capture
- starvation
- throwing harmful substances such as hot water, vinegar, acid or chemicals
- intentionally running over a dog
- explicit threats to kill, poison, beat, injure or otherwise physically harm dogs

An allegation that a vehicle deliberately targeted or intentionally ran over a dog belongs here.Do not assign physical_cruelty merely because a complainant 
uses emotional words such as "cruel", "heartless" or "brutal". The complaint must describe or explicitly allege intentional physical harm.

Example:

"The residents have been mistreating dogs. One resident intentionally ran over a small puppy."

physical_cruelty = 1

2. removal_or_relocation

Assign 1 when dogs are allegedly:

- captured
- picked up
- removed
- relocated
- displaced
- dumped
- abandoned elsewhere
- prevented from returning to their original location
- threatened with any of the above

Routine temporary capture for sterilisation or treatment should not automatically be classified as removal or relocation if 
the dogs were returned to the same location and no wrongful removal is alleged.

3. institutional_neglect

Assign 1 for alleged neglect, mistreatment or inadequate conditions while dogs are in the custody or care of an institution, including but not limited to:

- animal shelters
- veterinary hospitals or facilities
- ABC centres
- animal-welfare NGOs
- dog-catching agencies
- municipal facilities
- contractors responsible for animal care

Includes:

- overcrowding
- starvation
- lack of food or water
- unhygienic conditions
- mixing healthy dogs with infected dogs
- untreated illness, infection or injury
- improper sterilisation or treatment
- inadequate post-operative care
- death allegedly caused by institutional neglect or poor conditions

Do not assign institutional_neglect merely because the complaint disagrees with sterilisation in general. 
There must be an allegation of improper treatment, neglect, unsafe conditions, injury or inadequate care.

4. vehicle_collision

Assign 1 when a stray or community dog is accidentally or negligently hit, run over, injured or killed by a vehicle and intentional targeting is not clearly alleged.

Includes:

- hit-and-run
- careless driving
- reckless driving
- speeding
- failure to stop after hitting a dog
- accidental running over

For the same vehicle incident:

- clear intentional targeting → physical_cruelty = 1 and vehicle_collision = 0
- accidental, careless, reckless or negligent driving → vehicle_collision = 1 and physical_cruelty = 0
Do not infer intention merely because the driver failed to stop or behaved without remorse.
Example:

"A person hit a stray dog with his car, ran over it and did not stop. He later claimed that he did not see the dog."

vehicle_collision = 1
physical_cruelty = 0

5. other_harm

Assign 1 when the complaint clearly alleges harm to a dog, but that harm does not fit any of the categories above.
Do not use other_harm merely because the complaint contains additional background, legal arguments or feeder harassment.
Use other_harm only when a distinct form of harm to a dog is clearly alleged but the type of harm does not fall into the given categories.

6. unclear

Assign 1 when the complaint_category indicates harm to dogs, but the complaint text does not provide enough information to determine what harm occurred.
Examples may include:

- general opposition to sterilisation without an allegation of improper care
- descriptions of dogs dying from weather or disease without clearly alleging that a person or institution caused or contributed to the deaths

If physical_cruelty, removal_or_relocation, institutional_neglect, vehicle_collision or other_harm is 1, unclear must be 0.

--------------------------------------------------
PRIMARY ACTOR
--------------------------------------------------

Return exactly one primary_actor.

Allowed values:

- resident_group_or_rwa
- individual
- government_or_public_body
- abc_shelter_or_contractor
- private_institution_or_business
- unknown_or_unclear

Use the actor most directly alleged to have initiated, ordered or authorised the harm.
If the initiator or authorising actor is not clear, use the actor alleged to have directly carried it out.
Do not classify an authority merely because the complainant asks that authority to intervene. The authority must itself be accused of committing, ordering, authorising, facilitating or knowingly enabling the harm.

1. resident_group_or_rwa

Use for:

- Resident Welfare Associations
- housing societies
- apartment associations
- estate-management committees
- organised resident groups
- residents or neighbours described as acting collectively
- society security guards acting on behalf of the RWA or housing society

Examples:

- "Residents threatened to poison the dogs."
- "The neighbours collectively removed the puppies."
- "RWA members ordered the dogs to be taken away."
- "The housing society instructed security guards to beat the dogs."

Do not require the complaint to use the formal term "RWA". A collective of residents or neighbours can be coded here.

2. individual

Use for one or more identifiable private individuals acting personally rather than as an organised resident group or institution.

Examples:

- a named neighbour acting alone
- three identified men torturing a dog
- an individual driver using a personal vehicle
- a private person abducting or beating dogs

If the complaint refers broadly to "residents", "neighbours", "society members" or a collective colony action, use resident_group_or_rwa instead.
Do not assign individual merely because an act such as beating is described. The complaint must identify or clearly refer to a private person or persons as the actor.

3. government_or_public_body

Use for:

- municipal corporations
- panchayats
- police
- government departments
- government veterinary departments
- development authorities
- the armed forces
- government officials allegedly using official power or authority

Example:

"A panchayat ordered the dogs to be relocated."

If a government officer is accused specifically of using official influence, authority or position to cause the harm, use government_or_public_body.

4. abc_shelter_or_contractor

Use for an actor performing an animal-care, capture, sterilisation, impoundment, treatment or shelter role.

Includes:

- ABC programme operators
- animal shelters
- animal-welfare NGOs
- veterinary facilities
- dog-catching agencies
- municipal or private ABC contractors
- shelter or ABC staff
- agencies responsible for post-operative care

Use this category based on the actor's role in the incident, even when the operator is an NGO or contractor.

Example:

"An ABC operator picked up puppies, sterilised them improperly and kept them with infected dogs."

5. private_institution_or_business

Use for:

- schools
- colleges
- hospitals
- hotels
- shops
- builders
- construction companies
- factories
- employers
- private security agencies
- other private organisations

A person being described by their occupation or business ownership does not automatically make the actor a 
private institution. Use private_institution_or_business only when the harmful act is carried out through, 
on behalf of, or in the operational context of the business.

Example:

"Construction-company lorry drivers repeatedly ran over dogs, and management failed to act."

6. unknown_or_unclear

Use when the actor is not stated or when the complaint does not provide enough information to identify the appropriate actor category.

The actor can still be identifiable even when the type of harm is unclear.

--------------------------------------------------
ACTOR EDGE CASE
--------------------------------------------------

Complaint:

"The RWA instructed a private dog-catching agency to remove the dogs."

The RWA initiated and ordered the action.

primary_actor = resident_group_or_rwa

--------------------------------------------------
OUTPUT RULES
--------------------------------------------------

Return only one valid JSON object.

Do not include markdown.
Do not include an explanation.
Do not include any text before or after the JSON.

For eligible complaint categories, use this exact structure:

{{
  "physical_cruelty": 0,
  "removal_or_relocation": 0,
  "institutional_neglect": 0,
  "vehicle_collision": 0,
  "other_harm": 0,
  "unclear": 0,
  "primary_actor": "unknown_or_unclear"
}}

For an ineligible complaint category, return:

{{
  "physical_cruelty": "Not Applicable",
  "removal_or_relocation": "Not Applicable",
  "institutional_neglect": "Not Applicable",
  "vehicle_collision": "Not Applicable",
  "other_harm": "Not Applicable",
  "unclear": "Not Applicable",
  "primary_actor": "Not Applicable"
}}

Complaint category:
{complaint_category}
Complaint:
{complaint}
"""
# The values are supplied when you call the function. When supplied, it inserts those values here {complaint_category} and {complaint}
```

```python
# AI classification function
def ask_ai_to_classify(complaint, complaint_category, model):

    prompt = make_prompt(
        complaint=complaint,
        complaint_category=complaint_category
    )

    response = client.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )

    return response.choices[0].message.content
```

```python
#choosing model 
model = "openai/gpt-4.1"
```

```python
handcoded.columns
```

```python
# iterating over each row of handcoded dataset and running prompt
for idx, row in handcoded.iterrows():  # iterate over each complaint

    complaint = row["complaint"]  # this is where we tell the make prompt function what complaint and complaint_category is
    complaint_category = row["ai_complaint_type"]

    try:  # continue to the next row if one fails

        ai_answer = ask_ai_to_classify(
            complaint=complaint,
            complaint_category=complaint_category,
            model=model
        )

        parsed = json.loads(ai_answer)

        handcoded.loc[idx, "ai_physical_cruelty"] = parsed["physical_cruelty"]
        handcoded.loc[idx, "ai_removal_or_relocation"] = parsed["removal_or_relocation"]
        handcoded.loc[idx, "ai_inst_neglect_inadq_conditions"] = parsed["institutional_neglect"]
        handcoded.loc[idx, "ai_vehicle_collision_negli_driving"] = parsed["vehicle_collision"]
        handcoded.loc[idx, "ai_other_harm"] = parsed["other_harm"]
        handcoded.loc[idx, "ai_unclear"] = parsed["unclear"]
        handcoded.loc[idx, "ai_primary_actor"] = parsed["primary_actor"]

        print(f"Done row {idx}")

    except Exception as e:
        print(f"Error on row {idx}: {e}")
        print(ai_answer if "ai_answer" in locals() else "No AI answer")

    time.sleep(1)
```
